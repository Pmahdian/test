```python
import numpy as np
import seaborn as sns; sns.set()
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y_true = make_blobs(n_samples=800, centers=4, cluster_std=0.50, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], s= 10)
```

```python
model = KMeans(n_clusters=4)
model.fit(X)
y = model.predict(X)
```

```python
plt.scatter(X[:,0], X[:, 1],c = y, cmap='summer', s = 15)
centers = model.cluster_centers_
plt.scatter(centers[:,0], centers[:,1],c = 'red', s = 50, alpha=1)
```

```python
plt.scatter(X[:,0], X[:, 1],c = y_true, cmap='summer', s = 15)
centers = model.cluster_centers_
plt.scatter(centers[:,0], centers[:,1],c = 'red', s = 50, alpha=1)
```
